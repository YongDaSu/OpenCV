## p14有一些CAP相關參數

```python
#cap = cv2.VideoCapture(0, cv2.CAP_DSHOW) #0:指定webcam
cap = cv2.VideoCapture('vtest.avi') #直接用影片
#cap = cv2.VideoCapture('slow_traffic_small.mp4')

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
codec = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', codec, 20.0, (width, height))
#                                  編碼形式 fps
```

參數cv2.CAP_DSHOW是微軟特有的，用於防止釋放攝像頭時的[ WARN:0]，terminating async callback警告

```python
print(cap.isOpened())  #True
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
        #out.write(frame) #save color video
        
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #把彩色轉成灰階影片
        cv2.imshow('gray video', frame)
       
            
            
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
    else:
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
