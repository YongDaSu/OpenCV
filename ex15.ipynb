```python
import cv2 as cv
```

```python
def nothing(x):
    print(x)

#creat a window
cv.namedWindow('image')

cv.createTrackbar('CP', 'image', 0, 255, nothing) #current position
```

```python
switch = 'color/gray'
cv.createTrackbar(switch, 'image', 0, 1, nothing)
```

```python
while(1):
    img = cv.imread('lena.jpg')
    pos = cv.getTrackbarPos('CP', 'image')
    font = cv.FONT_HERSHEY_SIMPLEX
    cv.putText(img, str(pos), (100,150), font, 6, (217,25,255), 10)
    #cv2.putText(影像, 文字, 座標, 字型, 大小, 顏色, 線條寬度, 線條種類(預設LINE_AA))
    
    k = cv.waitKey(1) &0xFF
    if k == 27:
        break
        
    s = cv.getTrackbarPos(switch, 'image')
    
    if s == 0:
        pass
    else:
        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
        
    img = cv.imshow('image', img)
    
cv.destroyAllWindows()
```
