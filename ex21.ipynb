# Morphological Transformations

cv.erode()    
cv.dilate()     
cv.morphologyEx()    
Erosion: A pixel in the original image (either 1 or 0) will be considered 1 only if
all the pixels under the kernel is 1, otherwise it is eroded (made to zero).    
Dilation: a pixel element is ‘1’ if at least one pixel under the kernel is ‘1’.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('smarties.png', cv2.IMREAD_GRAYSCALE) # 0
plt.imshow(img)
_, mask = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY_INV)
#plt.imshow(mask)
#print(_)
```

```python
kernel = np.ones((5,5), np.uint8) # 5x5
dilation = cv2.dilate(mask, kernel, iterations=2)
erosion = cv2.erode(mask, kernel, iterations=1)
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
```

```python
titles = ['image', 'mask', 'dilation', 'erosion', 'opening', 'closing']
images = [img, mask, dilation, erosion, opening, closing]
```

```python
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
