```python
import numpy as np
import cv2 as cv
```

```python
cap = cv.VideoCapture('slow_traffic_small.mp4')
ret, frame = cap.read()

x, y, width, height = 300, 200, 100, 50
track_window = (x, y, width, height)

roi = frame[y:y+height, x:x+width]
hsv_roi = cv.cvtColor(roi, cv.COLOR_BGR2HSV)

mask = cv.inRange(hsv_roi, np.array((0., 60., 32.)), np.array((180., 255., 255)))
#mask.shape = (50, 100)

roi_hist = cv.calcHist([hsv_roi], [0], mask, [180], [0,180])

cv.normalize(roi_hist, roi_hist, 0, 255, cv.NORM_MINMAX)

term_crit = (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 10, 1)
cv.imshow('roi', roi)
#cv.imwrite('template2.jpg',roi)
```

```python
#roi2 = cv.imread('green.jpg')
#print(roi2.shape)


#x2, y2, width2, height2 = 50, 70, 30, 50
x2, y2, width2, height2 = 165, 130, 15, 30
track_window2 = (x2, y2, width2, height2)

roi2 = frame[y2:y2+height2, x2:x2+width2]
hsv_roi2 = cv.cvtColor(roi2, cv.COLOR_BGR2HSV)

#mask2 = cv.inRange(hsv_roi2, np.array((15., 30., 15.)), np.array((90., 200., 90.)))
mask2 = cv.inRange(hsv_roi2, np.array((20., 30., 20.)), np.array((90., 255., 90.)))
roi_hist2 = cv.calcHist([hsv_roi2], [0], mask2, [180], [0,180])

cv.normalize(roi_hist2, roi_hist2, 0, 255, cv.NORM_MINMAX)

term_crit2 = (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 10, 1)
cv.imshow('roi2', roi2)
```

```python
while(1):
    ret, frame = cap.read()
    if ret == True:
        
        hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
        dst = cv.calcBackProject([hsv], [0], roi_hist, [0,180], 1)
        
        dst2 = cv.calcBackProject([hsv], [0], roi_hist2, [0,180], 1)
        
        ret, track_window = cv.CamShift(dst, track_window, term_crit)
        
        ret2, track_window2 = cv.CamShift(dst2, track_window2, term_crit2)
        
        pts = cv.boxPoints(ret) #找出矩形4個頂點
        print(pts)
        pts = np.int0(pts)
        
        pts2 = cv.boxPoints(ret2) #找出矩形4個頂點
        print(pts2)
        pts2 = np.int0(pts2)

        final_image = cv.polylines(frame, [pts], True, (0,255,255), 2) #非填充多邊形
        final_image2 = cv.polylines(final_image, [pts2], True, (0,255,255), 2) #非填充多邊形
        
        cv.imshow('dst', dst)
        cv.imshow('final_image', final_image2)
        k = cv.waitKey(30) & 0xFF
        if k == 27:
            break
    
    else:
        break
cap.release()
cv.destroyAllWindows()
```
